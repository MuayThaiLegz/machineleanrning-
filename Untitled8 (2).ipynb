```python
# reading data in, droping Unnamed column and setting both case_id and phenotype to index

Shutdown_Physiologic = pd.read_csv("Shutdown_Physiologic.csv").drop('Unnamed: 0', axis=1).set_index(['case_id','phenotype'])


#dummy_data  = pd.DataFrame(Shutdown_Physiologic.values,  columns=Shutdown_Physiologic.columns)

#Shutdown_Physiologic = pd.merge(Shutdown_Physiologic , dummy_data[::-1])

# Using OneHotEncoder to encode our dependent variable into a few categories:


cat_encoder = OneHotEncoder(sparse=True)

pheno_type_cat = Shutdown_Physiologic[['pheno']]
pheno_type_cat_1hot = cat_encoder.fit_transform(pheno_type_cat)

#Shutdown_Physiologic['pheno_physiological'], Shutdown_Physiologic['pheno_shutdown'] = pheno_type_cat_1hot.toarray()[:, 1], pheno_type_cat_1hot.toarray()[:, 0]

Shutdown_Physiologic['pheno'] = pheno_type_cat_1hot.toarray()

y_ = Shutdown_Physiologic['pheno']
#y_pheno_shutdown = Shutdown_Physiologic[['pheno_shutdown']]
#y_pheno_physiological = Shutdown_Physiologic[['pheno_physiological']]

X_ = Shutdown_Physiologic.drop(["pheno"],axis=1)



#cor = Shutdown_Physiologic[filterlabels.index].corr() 


#plt.figure(figsize=(12,10))
#sns.heatmap(cor)
#plt.title('Heatmap for the Data Correlation')
#plt.show()
```

```python
mi = mutual_info_classif(X=X_,y=y_,random_state=42)    # Estimate mutual information for a discrete target variable. Which measures dependecy between variables  
dfinfo = pd.DataFrame(mi, index=X_.columns) # Creating dateframe from results
filterlabels = dfinfo[dfinfo[0] >= 0.09] # Filtering resulting dateframe

filterlabels
```

```python


Mut_info  = Shutdown_Physiologic[filterlabels.index]
MI_X_train, MI_X_test, MI_y_train, MI_y_test = train_test_split(X_[filterlabels.index], y_, stratify=y_)

# Assess model
lr = LogisticRegression()
lr.fit(MI_X_train, MI_y_train)
strat_kfold = StratifiedKFold(5)
score = cross_val_score(lr, MI_X_train, MI_y_train, scoring='accuracy', cv=10)
print('CV accuracy scores: %s' % score)

print('CV accuracy: %.3f +/- %.3f' % (np.mean(score), np.std(score)))
```

```python


svm_clf = SVC(random_state=42)
log_clf = LogisticRegression(random_state=42)
rnd_clf = RandomForestClassifier(random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf),
                ('svc', svm_clf)],
    voting='hard')
voting_clf.fit(MI_X_train, MI_y_train)


for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(MI_X_train, MI_y_train)
    y_pred = clf.predict(MI_X_test)
    print(clf.__class__.__name__, accuracy_score(MI_y_test, y_pred))
```

```python
from sklearn.cluster import KMeans


km = KMeans(n_clusters=4, 
            init='random', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)

y_km = km.fit_predict(X)

# Kmean clustering shows the high dimentinality of our data

# We are looking to plot in 3d and also reduce dimentionality as needed

plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='lightgreen',
            marker='s', edgecolor='black',
            label='Cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='Cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='Cluster 3')
plt.scatter(X[y_km == 3, 0],
            X[y_km == 3, 1],
            s=50, c='lightblue',
            marker='x', edgecolor='black',
            label='Cluster 4')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='Centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
def threeDPlot(data):
        z = data.rating
        x = data.building
        y = data.category_int
 
# Creating figure
        fig = plt.figure(figsize = (10, 8))
        ax = plt.axes(projection ="3d")
   
# Add x, y gridlines
        ax.grid(b = True, color ='black',
                linestyle ='--', linewidth = 0.3,
                )
 
 
# Creating color map
        my_cmap = plt.get_cmap('coolwarm')
 
# Creating plot
        sctt = ax.scatter3D(x, y, z,
                #    alpha = .9,
                    c = z,
                    cmap = my_cmap,
                    marker ='*')
 
#plt.title(' | '.join(restaurant_data_dfname))
        ax.set_xlabel('building', fontweight ='bold')
        ax.set_ylabel('category_int', fontweight ='bold')
        ax.set_zlabel('rating', fontweight ='bold')
        fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
 
# show plot
        plt.show()
```

```python
from sklearn.decomposition import PCA

def PCA_ns(n=int):
  pca = PCA(n)
  pca.fit(X)
  pca.explained_variance_ratio_.sum()

  plt.bar(range(n), pca.explained_variance_ratio_, alpha=0.5, align='center')
  plt.step(range(n), np.cumsum(pca.explained_variance_ratio_), where='mid')
  plt.ylabel('Explained variance ratio')
  plt.xlabel('Principal components')
  plt.title(f'The explained variance ratio is: {pca.explained_variance_ratio_.sum()}')

  plt.show()
  return pd.DataFrame(pca.components_)
```
