```python
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
import seaborn as sns
import pandas as pd
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split

import numpy as np

gpu_devices = tf.config.experimental.list_physical_devices('GPU')

if gpu_devices:
    print('Using GPU')
    tf.config.experimental.set_memory_growth(gpu_devices[0], True)
else:
    print('Using CPU')


pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# reading data in, droping Unnamed column and setting both case_id and phenotype to index

Shutdown_Physiologic = pd.read_csv("Shutdown_Physiologic.csv").drop('Unnamed: 0', axis=1).set_index(['case_id','phenotype'])


dummy_data  = pd.DataFrame(Shutdown_Physiologic.values,  columns=Shutdown_Physiologic.columns)

Shutdown_Physiologic = pd.merge(Shutdown_Physiologic , dummy_data[::-1])

```

```python

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import OneHotEncoder
```

```python
# Using OneHotEncoder to encode our dependent variable into a few categories:
s

cat_encoder = OneHotEncoder(sparse=True)

pheno_type_cat = Shutdown_Physiologic[['pheno']]
pheno_type_cat_1hot = cat_encoder.fit_transform(pheno_type_cat)

#Shutdown_Physiologic['pheno_physiological'], Shutdown_Physiologic['pheno_shutdown'] = pheno_type_cat_1hot.toarray()[:, 1], pheno_type_cat_1hot.toarray()[:, 0]

Shutdown_Physiologic['pheno'] = pheno_type_cat_1hot.toarray()
```

```python

#Shutdown_Physiologic[['pheno','pheno_physiological','pheno_shutdown']].head(30)
```

```python
y_ = Shutdown_Physiologic[['pheno']]
y_pheno_shutdown = Shutdown_Physiologic[['pheno_shutdown']]
y_pheno_physiological = Shutdown_Physiologic[['pheno_physiological']]

X_ = Shutdown_Physiologic.drop(["pheno",'pheno_physiological','pheno_shutdown'] ,axis=1)

y = np.asarray(y_.values)
X = np.asarray(X_.values)



X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
from sklearn.cluster import KMeans


km = KMeans(n_clusters=4, 
            init='random', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)

y_km = km.fit_predict(X)

# Kmean clustering shows the high dimentinality of our data

# We are looking to plot in 3d and also reduce dimentionality as needed

plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='lightgreen',
            marker='s', edgecolor='black',
            label='Cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='Cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='Cluster 3')
plt.scatter(X[y_km == 3, 0],
            X[y_km == 3, 1],
            s=50, c='lightblue',
            marker='x', edgecolor='black',
            label='Cluster 4')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='Centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
distortions = []
for i in range(1, 7):
    km = KMeans(n_clusters=i, 
                init='k-means++', 
                n_init=10, 
                max_iter=300, 
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 7), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
#plt.savefig('images/11_03.png', dpi=300)
plt.show()
```

```python
# creating scatter plots to compare columns with neighboring column looking to extract insight
for i in range(52):
  plt.scatter(X[:, i], X[:, i+1], 
            c='grey', marker='o', edgecolor='blue', s=50)
  plt.grid()
  plt.tight_layout()
  plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif 
import seaborn as sns


mi = mutual_info_classif(X=X_,y=y_,random_state=42)    # Estimate mutual information for a discrete target variable. Which measures dependecy between variables
  
dfinfo = pd.DataFrame(mi, index=X_.columns) # Creating dateframe from results

filterlabels = dfinfo[dfinfo[0] >= 0.10] # Filtering resulting dateframe

# Using filtered results to create a corrlation heatmap
cor = Shutdown_Physiologic[filterlabels.index].corr()
plt.figure(figsize=(12,10))
sns.heatmap(cor)
plt.title('Heatmap for the Data Correlation')
plt.show()
```

```python
from sklearn.decomposition import PCA


pca = PCA(20)
pca.fit(X)
pca.explained_variance_ratio_.sum()

plt.bar(range(20), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(20), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.title(f'The explained variance ratio is: {pca.explained_variance_ratio_.sum()}')

plt.show()
```

```python
pca_components = pd.DataFrame(pca.components_)
```

```python
# Using filtered results to create a corrlation heatmap
cor_pca_components = pca_components.corr()
plt.figure(figsize=(12,10))
sns.heatmap(cor_pca_components)
plt.title('Heatmap for the Data Correlation')
plt.show()
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score


xgb_clf = XGBClassifier(max_depth=3,                  # Maximum tree depth for base learners.
                        learning_rate=0.1,            # Boosting learning rate (xgb's "eta")
                        n_estimators=100,             # Number of boosted trees to fit.
                        silent=True,                  # Whether to print messages while running
                        objective='binary:logistic',  # Task and objective or custom objective function
                        booster='gbtree',             # Select booster: gbtree, gblinear or dart
#                         tree_method='gpu_hist',
                        n_jobs=-1,                    # Number of parallel threads
                        gamma=0.1,                      # Min loss reduction for further splits
                        min_child_weight=1,           # Min sum of sample weight(hessian) needed
                        max_delta_step=0,             # Max delta step for each tree's weight estimation
                        subsample=1,                  # Subsample ratio of training samples
                        colsample_bytree=1,           # Subsample ratio of cols for each tree
                        colsample_bylevel=1,          # Subsample ratio of cols for each split
                        reg_alpha=0,                  # L1 regularization term on weights
                        reg_lambda=1,                 # L2 regularization term on weights
                        scale_pos_weight=1,           # Balancing class weights
                        base_score=0.5,               # Initial prediction score; global bias
                        random_state=42)   


scores = cross_val_score(estimator=xgb_clf,
                         X=X_train,
                         y=y_train,
                         cv=5,
                         n_jobs=1)

print('CV accuracy scores: %s' % scores)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```
